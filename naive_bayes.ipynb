```python
import pandas as pd
df=pd.read_csv('../venv_22/titanic.csv')
df
```

```python
df.info()
```

```python
df.Age=df.Age.fillna(df.Age.mean())
```

```python
df.info()
```

```python
target=df.Survived
target
```

```python
dummies=pd.get_dummies(df.Sex)
dummies
```

```python
df=pd.concat([df,dummies],axis='columns')
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived','Sex'],axis='columns',inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.4, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

### Spam

```python
sp=pd.read_csv('../venv_22/spam.csv')
sp
```

```python
sp.groupby('Category').describe()
```

```python
sp['spam']=sp['Category'].apply(lambda x: 1 if x=='spam' else 0)
sp.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sp.Message,sp.spam)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
